```python
graph = nx.DiGraph()

graph.add_nodes_from(['A','B','C','D'])
graph.add_edges_from([('A','B'), ('B','C'), ('C', 'D'), ('D', 'A')])

pos = {'A': (1, 1), 'B': (1, 4), 'C': (4, 4), 'D': (4, 1)}

def draw_G(G, axis=False, pos=None):
    options = {
        "font_size": 20,
        "node_size": 1500,
        "node_color": "white",
        "edgecolors": "black",
        "linewidths": 3,
        "width": 3,
    }

    if axis:
        plt.axis([0, 6, -0.5, 6])
    nx.draw_networkx(G, pos, **options)
    plt.show()
    
draw_G(graph, True, pos)
```
